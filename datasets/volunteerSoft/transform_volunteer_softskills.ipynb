```python
df_.set_index("TASK #", inplace=True)
df_
```

```python
#df_.reset_index(inplace=True)
skill_cols = df_.columns.difference(['Task ID', 'SHORT'])
def list_non_nan_skills(row, cols=skill_cols):
    """Gibt alle Spaltennamen zurück, deren Wert in der Zeile nicht NaN ist."""
    return row[cols].dropna().index.tolist()
df_['ESCO_list'] = df_.apply(list_non_nan_skills, axis=1)
df_.reset_index(drop=True, inplace=True)
#df_.reset_index()
df_
```

```python
df_ = df_[["Task ID", "SHORT", "ESCO_list"]]

# reanme cols
df_.rename(columns={"Task ID": "task_id", "SHORT": "short", "ESCO_list": "y"}, inplace=True)
df_
```
