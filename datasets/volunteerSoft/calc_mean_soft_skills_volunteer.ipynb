```python
from ast import literal_eval

import pandas as pd
df__ = pd.read_csv("volunteer_softskills_16_05.csv", sep=";", index_col=0)
df__.T
```

```python
from datasets import volunteerDataset
```

```python
volunteerDataset.df
```

```python
# add column "final_cause_area" from volunteerDataset.df to df__ on "task_id"
df__ = pd.merge(df__, volunteerDataset.df[["task_id", "final_cause_area"]], on="task_id", how="left")
```

```python
for i, row in df__.iterrows():
    new_list = []
    for skill in literal_eval(row['y']):
        if " (***NEU)" in skill:
            print(f"Replacing {skill} with {skill.replace(' (***NEU)', '')}")
            skill = skill.replace(" (***NEU)", "")
            new_list.append(skill)
        elif "\n" in skill:
            print(f"Removing newline from {skill}")
            skill = skill.replace("\n", "")
            new_list.append(skill)
        else:
            new_list.append(skill)

    # update the row of df to new list
    df__.at[i, 'y'] = new_list
```

```python
# save df__ to csv
df__.to_csv("volunteer_softskills_16_05.csv", sep=";", index=True)
```
